

Use dot product to calculate the profit for each day

```python
import numpy as np
import pandas as pd

sales = np.array([[2,7,1],[9,4,16],[11,14,18],[13,13,16],[15,18,9]])
price_per_unit = np.array([[10], [8], [12]])

profit = np.dot(sales, price_per_unit)


print("Size of sales", sales.shape)
print("Size of price_per_unit", price_per_unit.shape)
print("Mon = ", profit[0])
print("Tues = ", profit[1])
print("Wed = ", profit[2])
print("Thu = ", profit[3])
print("Fri = ", profit[4])
```

### Attempt 2 - use pandas and numpy

```python
weekly_sales_table = pd.DataFrame(sales, index=["Mon", "Tue", "Wed", "Thu",  "Fri"], columns=["Almond", "Peanut", "Cashew butter"])
price_unit_table = pd.DataFrame(price_per_unit, index=["Almond", "Peanut", "Cashew butter"], columns=["Price"])
profit = pd.DataFrame(np.dot(weekly_sales_table, price_per_unit), index=["Mon", "Tue", "Wed", "Thu",  "Fri"], columns=["Profit"])

print(weekly_sales_table)
print("----------------------------------------")
print(price_unit_table)
print("----------------------------------------")
print(profit)



```
