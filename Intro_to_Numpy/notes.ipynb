## Creating array

- np.array([element])  - to create ndarray with manual input

`Automated`
- np.one((size))                  - create ndarray fill with 1s (float64)
- np.arange(start, stop , step)   - create ndarray filled a pattern
- np.random.randint(0, 10, size=(3,5)) - create ndarray random  

`Random seed - make random numbers produceable`
- np.random.seed(seed=0)

```python
print(f"------------ Manual -------------")
print(np.array([1, 2, 3]))
print(f"------------ Fill with ones -------------")
print(np.ones((5,1,2)))
print(f"------------ Fill with pattern -------------")
print(np.arange(3, 12 , 2))
print(f"------------ Fill size with random num -------------")
print(np.random.randint(0, 10, size=(3,5)))
print(f"------------ Fill size with random with seed -------------")
np.random.seed(seed=1)
print(np.random.randint(0, 10, size=(3,5)))
```

## View Arrays and Matrices

np.unique(ndarray)  - show uniques elements

[ ,  ,]   - Select indices just like index array
[: , :, :] - Also can used slices

```python
print(np.unique(np.random.randint(0, 10, size=(3,5))))
print("-----------------------")

test = np.random.randint(0, 10, size=(3,5,4))
print(test)
print("---------  Access ndarray --------------")
print(test[2,0, 1])
print("---------  Access ndarray --------------")
print(test[:2,:2, :1])
                 # <------ View in reverse
```
