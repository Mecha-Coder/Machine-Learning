```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## Plot workflow


```python
#1. Prepare data
x = np.array([1,2,3,4])
y = np.array([11, 22, 33, 44])

#2. Setup
fig, ax = plt.subplots(figsize=(2, 2))

#3. Plot data
ax.plot(x,y)

#4. Customize plot
ax.set(title="Simple Plot",
       xlabel="x-axis",
       ylabel="y-axis",
       )

#5. Save & show (you save the whole figure)
fig.savefig("export_img.png")
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)  # 1 row, 2 columns
ax1.plot(x, y)      # Plot on left
ax2.plot(x, y)      # Plot on right

```

```python
fig, ax = plt.subplots() # setup plot
ax.plot(x, 2 ** x) # Add data
```

```python
# Data
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 5, 4, 6])

# Create scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x))  # Use .scatter() instead of .plot()
```

```python
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x))
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10,5))
ax1.plot(x, x/2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(y, x)
ax4.hist(np.random.random(1000))
ax1.set(title="Title", ylabel="Price")
```

```python
# or 

fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10,5))
ax[0,0].plot(x, x/2)
ax[0,1].scatter(np.random.random(10), np.random.random(10))
ax[1,0].bar(y, x)
ax[1,1].hist(np.random.random(1000))
```
