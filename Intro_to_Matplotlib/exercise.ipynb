```python
# 14) Python dictionary of 3 of your favourite foods. Plot a horizontal and vertical bar 


food = {
    "Apple" : 10,
    "Banana": 5,
    "Eggs": 6
}

fig, ax = plt.subplots(1,2, figsize=(11, 4))

ax[0].bar(food.keys(), food.values())
ax[0].set(title="Fruits", xlabel="Item", ylabel="Price($)")

ax[1].barh(food.keys(), food.values())
ax[1].set(title="Fruits", xlabel="Item", ylabel="Price($)")
```

```python
# Plot using pandas
import pandas as pd

car_sales = pd.read_csv("car-sales.csv")
car_sales["Price"] = car_sales["Price"].str.replace(r"[\$\,\.]", "", regex=True)
car_sales["Price"] = car_sales["Price"].str[:-2]


# All accumulative sales
car_sales["Total Sales"] = car_sales["Price"].astype(int).cumsum()
car_sales["Sale Date"] = pd.date_range("13/1/2020", periods=len(car_sales))
car_sales
```

```python
car_sales.plot(x="Sale Date", y="Total Sales", marker='o', linewidth=5, color="red", alpha=0.2, linestyle="--")
car_sales
```

```python
car_sales["Price"] = car_sales["Price"].astype(int)
car_sales.plot(x="Odometer (KM)", y="Price", kind="scatter")
```

```python
heart_disease = pd.read_csv("heart-disease.csv")
heart_disease
heart_disease["age"].plot(kind="hist",bins=50)
```

```python
heart_disease.plot.hist(figsize=(10, 30), subplots=True);
```

```python
over_50 = heart_disease[ heart_disease["age"] > 50]
over_50


```

```python

fig, ax = plt.subplots(figsize=(10, 8))
scatter = ax.scatter(
    over_50["age"],
    over_50["chol"],
    c=over_50["target"],   ### Scatter have this property called c
    cmap="winter",
)

scatter = ax.legend(*scatter.legend_elements(), title="Target")

ax.set(
    title="Heart Disease and Choleesterol Levels",
    xlabel="Age",
    ylabel="Cholestrol"
)

ax.axhline(over_50["chol"].mean(), 
    linestyle="--",
    color="red"
    )

```
